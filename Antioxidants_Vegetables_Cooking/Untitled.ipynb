```python
# Covid tests, entropy and information 
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p = np.arange(1,10,0.01)
p = np.exp(p)
p = p/p.max()
```

```python
plt.plot(p)
```

```python
def E(p):
    return -(p*np.log(p)  + (1-p)*np.log(1-p))
```

```python
e = E(p)
```

```python
ax = plt.subplot()
plt.plot(p,e)
ax.set_xscale('log')
```
